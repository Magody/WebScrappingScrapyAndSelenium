```python
import pandas as pd

df = pd.read_csv("merged_shards.csv")
```

```python
df.columns
```

```python
mask = (df["reactions_total"]>0) & (df["state"] != "404 not found")
df_useful = df.loc[mask, :]
df_useful.head()
```

```python
df_useful["score_reaction"] = (df_useful["reactions_like"] + df_useful["reactions_love"])/df_useful["reactions_total"]
df_sorted = df_useful.sort_values(["rating", "votes", "score_reaction", "reactions_total"], axis=0,ascending=False)
```

```python
top100 = df_sorted.loc[:, ["title", "score_reaction", "rating", "votes"]].head(100)
top100.to_csv("top100animes.csv")
```
